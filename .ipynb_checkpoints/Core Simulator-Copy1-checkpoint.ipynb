```python
#library building
import networkx as nx
import math
import random as rand
import numpy as np
import copy
%matplotlib inline
import matplotlib.pyplot as plt
from itertools import combinations as comb
```

```python
#**** Graph building helper functions ****#

#degree k, rewiring probability p
def get_watts_strogatz_graph(nodeNum,k,p,):   
    G = nx.watts_strogatz_graph(nodeNum,k,p)
    for e in G.edges():
        G[e[0]][e[1]]['weight'] = 1
    return G

# Build g
def get_undirected_ER_graph(nodeNum, p):
    G = nx.erdos_renyi_graph(nodeNum,p)
    for e in G.edges():
        G[e[0]][e[1]]['weight'] = 1
    return G

#Generate tree with powerlaw degree distribution
def get_powerlaw_tree_graph(nodeNum,gamma):
    G = nx.random_powerlaw_tree(nodeNum,gamma,tries=10000)
    for e in G.edges():
        G[e[0]][e[1]]['weight'] = 1
    return G
```

```python
#Graph helper functions

def draw_graph_helper(H,positionFlag="spring"):
    
    if (positionFlag.startswith("spring")):
        pos=nx.spring_layout(H,iterations=20)
    if (positionFlag.startswith("random")):
        pos=nx.random_layout(H)
    plt.figure(figsize=(10,10))
    
    nx.draw_networkx_nodes(H,pos)
    nx.draw_networkx_edges(H,pos)
    nx.draw_networkx_labels(H,pos,fontsize=12)
    
    labels = nx.get_edge_attributes(H,'weight')
    labelsNonZero = {}
    for l in labels.keys():
        if labels[l] > 0.0001:
            labelsNonZero[l] = labels[l]
            
    
    nx.draw_networkx_edge_labels(H,pos,edge_labels=labelsNonZero)
    
    plt.show()
    return
```

```python
#Construct a graph whose edge weights are the likelihood of an awareness relationship, given the contact graph G
#under the JS spline metric model
# --> This is a symetric measure
def construct_awareness_from_contact_graph(G):
    H = G.copy()
    #print "Edges",H.edges()
    for pair in comb(G.nodes(),2): #get pairs
        x = pair[0]
        y = pair[1]
       
        metric = 0.0
        if (y in G[x].keys()):#if y a neighor of x (and visa versa)
            metric = 0.5
            
        #JS Metric
        xNbr = set(G[x])
        yNbr = set(G[y])
        mutualNbrs = xNbr.intersection(yNbr)
        
        metric = metric + len(mutualNbrs)/float(len(xNbr.union(yNbr)))
        
        #set as edge weight in new graph
        if (y not in H[x]):
            H.add_edge(x,y)
        H[x][y]['weight'] = metric
        
    return H
        
    
```

```python
#Calculate the expected number of neighbors of S in nodeSet, not including S itself
def get_exp_coverage(G,nodeSet,S):
    #Coverage probabilities of
    #print "Finding coverage of S = ",S
    covg = {}
    
    for n in S:
        #Get neighbors
        for b in G[n]:
            #If not in nodeSet or in S, ignore
            if (b not in nodeSet or b in S):
                continue
            #If in nodeSet and already covered by some node, update coverage
            elif (b in covg.keys()): 
                #Update coverage of b
                covg[b] = 1 - (1-covg[b])*(1-G[n][b]['weight'])
            else:
                covg[b] = G[n][b]['weight']
                
    #return the sum of all of the coverage weights
    return sum([covg[x] for x in covg.keys()])
    
# Find approximation max coverage set using k-greedy method, with t total nodes in solution
def greedy_expected_max_coverage_set(G,t,k):
    #Debug
    assert(t < len(G.nodes()))
    
        #solution
    soln = set()
           
    #node set
    remaining_nodes = set(G.nodes())
    
    #Construct how many nodes to select at each step
    k_vals = [k for _ in range(int(math.floor(float(t)/k)))]
    if (t - sum(k_vals) > 0):
        k_vals.append(t - sum(k_vals))
    #Debug
    assert(sum(k_vals) == t)
    
    for sub_k in k_vals:
        
        #Get value of current (partial solution)
        if (len(soln) > 0):
            current_covg = get_exp_coverage(G,G.nodes(),soln)
        else:
            current_covg = 0.0
           
        max_sym_diff = 0.0
        best_seen_soln = set()
        remaining_nodes = remaining_nodes.difference(soln)
        
        #print "\n Selecting next ", sub_k, "nodes from set ", remaining_nodes
        for nodeset in comb(remaining_nodes,sub_k): #For all k combinations of unselected nodes
            #calculate expected coverage with the addition of nodeset to the current solution
            sym_diff = get_exp_coverage(G,G.nodes(),soln.union(nodeset))
            #print "Coverage of ", soln.union(nodeset), "is ", sym_diff
            if (sym_diff > max_sym_diff):
                #print "Using ", nodeset, "as current best"
                max_sym_diff = sym_diff
                best_seen_soln = nodeset
        #print "Chose node(s) ", list(best_seen_soln)
        #Add found nodes to solution set
        #Debug
        assert(len(set(best_seen_soln).intersection(soln)) == 0)
        soln = soln.union(set(best_seen_soln))
    return soln,max_sym_diff
```

```python
#Given a contact graph G and a committee, form the clique subgraph of the committee and perform triadic closure
# using the closure param
# Note: this involves closing triads with three non-committee members
# Assumes that contact_graph is undirected

#Returns a new graph and the number of new edges added
def simple_committee_closure_augmentation(contact_graph, committee, closure_param):
    #pairs
    G = contact_graph.copy()
    
    for v,w in comb(committee,2):
        G.add_edge(v,w)
        
    #Closed neighborhood of committee
    nbrs = set(committee)
    for c in committee:
        nbrs = nbrs.union(set(G[c].keys()))
        
    for x,y,z in comb(nbrs,3):
        if (x in G[y] and z in G[y] and z not in G[x]):
            if (rand.random() < closure_param):
                G.add_edge(x,z)
        if (x in G[z] and y in G[z] and x not in G[y]):
            if (rand.random() < closure_param):
                G.add_edge(x,y)
        if (y in G[x] and z in G[x] and y not in G[z]):
            if (rand.random() < closure_param):
                G.add_edge(y,z)
                                                    
    #Count number of new edges
    new_edges = len(set(G.edges()).difference(set(contact_graph.edges())))
    
    return G,new_edges

# Close triads which include two members of the committee
#Returns a new graph and the number of new edges added
def committee_only_closure_augmentation(contact_graph, committee, closure_param):
    #pairs
    G = contact_graph.copy()
    
    for v,w in comb(committee,2):
        G.add_edge(v,w)
        
    #Closed neighborhood of committee
    nbrs = set(committee)
    for c in committee:
        nbrs = nbrs.union(set(G[c].keys()))
        
    for x,y,z in comb(nbrs,3):
        #dont close triads which lie entirely outside committee
        if (x not in committee and y not in committee and z not in committee):
            continue
        if (x in G[y] and z in G[y] and z not in G[x]):
            if (rand.random() < closure_param):
                G.add_edge(x,z)
        if (x in G[z] and y in G[z] and x not in G[y]):
            if (rand.random() < closure_param):
                G.add_edge(x,y)
        if (y in G[x] and z in G[x] and y not in G[z]):
            if (rand.random() < closure_param):
                G.add_edge(y,z)
                                                    
    #Count number of new edges
    new_edges = len(set(G.edges()).difference(set(contact_graph.edges())))
    
    return G,new_edges
    
    
```

```python
def dynamics_record():
    
    Y_covg_list = []
    Y_clust_list = []
    Y_ddist_list = []
    Y_decc_list = []
    Y_nedge_list = []
    Y_graph_list = []

    for closure_param in closure_param_list:
        #Store results for this closure param
        Y_covg_list.append([])
        Y_clust_list.append([])
        Y_ddist_list.append([])
        Y_decc_list.append([])
        Y_nedge_list.append([])
        Y_graph_list.append([])
        Y_covg = Y_covg_list[-1]
        Y_clust = Y_clust_list[-1]
        Y_ddist = Y_ddist_list[-1]
        Y_decc = Y_decc_list[-1]
        Y_nedge = Y_nedge_list[-1]
        Y_graph = Y_graph_list[-1]

        for t in range(TRIALS):
            #print "\nRunning trial ",t," for closure param",closure_param
            Y_covg.append([])
            Y_clust.append([])
            Y_ddist.append([])
            Y_decc.append([])
            Y_nedge.append([])
            Y_graph.append([])

            #build random graph of chosen type
            if (GRAPH_TYPE.startswith("PL_TREE")):
                contact_net = get_powerlaw_tree_graph(N,alpha)
            elif (GRAPH_TYPE.startswith("ER")):
                contact_net = get_undirected_ER_graph(N,er_param)
            elif (GRAPH_TYPE.startswith("WS")):
                contact_net = get_watts_strogatz_graph(N,ws_degree,ws_rewiring_prob)

            for i in range(iterations):
                #print "Itertation ", i, "starting;",
                Y_graph.append(contact_net)
                #Build awareness graph for covoerage
                awareness_graph = construct_awareness_from_contact_graph(contact_net)

                #Cvg saves committee, exp. coverage tuple
                cvg = greedy_expected_max_coverage_set(awareness_graph,committee_size,1) #greedy max coverage on awareness graph
                Y_covg[-1].append(cvg[1])


                #Save clustering distribution
                Y_clust[-1].append(nx.clustering(contact_net).values())

                #Save degree distribution
                Y_ddist[-1].append(nx.degree(contact_net).values())

                #Save eccentricities
                Y_decc[-1].append(nx.eccentricity(contact_net).values())

                # Apply dynamics, augment graph
                if (DYNAMICS_TYPE.startswith("full-closure")):
                    contact_net,new_edge_cnt = simple_committee_closure_augmentation(contact_net,cvg[0],closure_param)
                elif (DYNAMICS_TYPE.startswith("committee-only-closure")):
                    contact_net,new_edge_cnt = committee_only_closure_augmentation(contact_net,cvg[0],closure_param)
                else:
                    assert(0)
                Y_nedge[-1].append(new_edge_cnt)

            #add final graph
            Y_graph.append(contact_net)

    Y_all = (Y_covg_list,Y_clust_list,Y_ddist_list ,Y_decc_list ,Y_nedge_list,Y_graph_list) 
    return Y_all
```

```python
#Helper to plot aggregated histograms
def plot_aggregated_histograms(Y,num_aggregates,num_iterations,f,axis):
    Y_prime = [[] for i in range(iterations)]
    
    for t in range(num_aggregates):
        for i in range(num_iterations):
            Y_prime[i] += Y[t][i]
    #debug
            #print "Y[",t,"][",i,"] =", Y[t][i]
    #print Y_prime
    '''
    for i in range(iterations):
        binwidth = 1
        bins=np.arange(min(Y_prime[i]), max(Y_prime[i]) + binwidth, binwidth)
        ax.hist(Y_prime[i], bins=bins,alpha=0.5,label="iter %i"%i,color=np.random.rand(1,3))
    '''
      
    #easy version first
    binwidth = 1
    mn = np.floor(min([min(Y_prime[i]) for i in range(num_iterations)]))
    mx = max([max(Y_prime[i]) for i in range(num_iterations)])
    bins=np.arange(mn,  mx , (mx-mn)/20)
    plt.xticks(np.arange(mn,mx,(mx-mn)/20),rotation='vertical')
        
    #bins=np.arange(mn,  mx , 0.25)
    color=np.random.rand(3,num_iterations)
    #print Y_prime
    ax.hist(Y_prime,bins=bins,label=range(num_iterations))
    
    ax.legend()
```

```python
def helper_plot_stats(results_package):
    (Y_covg_list,Y_clust_list,Y_ddist_list ,Y_decc_list ,Y_nedge_list,Y_graph_list) = results_package
    for i,param in enumerate(closure_param_list):
        f,ax = plt.subplots(figsize=(10,10))
        title = "Degree distribution for closure param = %3f" % param
        plt.title(title)
        #print "Plotting aggregated degree distributions over time"
        #Take the ith param and get the list of lists
        plot_aggregated_histograms(Y_ddist_list[i],TRIALS,iterations,f,ax)
        f.tight_layout()
        plt.show()

    for i,param in enumerate(closure_param_list):    
        f,ax = plt.subplots(figsize=(10,10))
        title = "Clustering Distribution for closure param = %3f" % param
        plt.title(title)
        #print "Plotting aggregated degree distributions over time"
        #Take the ith param and get the list of lists
        #print Y_clust_list[i]
        plot_aggregated_histograms(Y_clust_list[i],TRIALS,iterations,f,ax)
        f.tight_layout()
        plt.show()

    for i,param in enumerate(closure_param_list):    
        f,ax = plt.subplots(figsize=(10,10))
        title = "Eccentricity Distribution for closure param = %3f" % param
        plt.title(title)
        #print "Plotting aggregated degree distributions over time"
        #Take the ith param and get the list of lists
        plot_aggregated_histograms(Y_decc_list[i],TRIALS,iterations,f,ax)
        f.tight_layout()
        plt.show()

    f,ax = plt.subplots(figsize=(10,10))
    plt.title("Coverage of Committees")
    plot_aggregated_trends(Y_covg_list,TRIALS,iterations,closure_param_list,f,ax)
    plt.show()
    f,ax = plt.subplots(figsize=(10,10))
    plt.title("(cumulative)Coverage of Committees")
    plot_aggregated_trends(Y_covg_list,TRIALS,iterations,closure_param_list,f,ax,cumulative=True)
    plt.show()

    print "Plotting new edges added over time in graph", GRAPH_TYPE
    f,ax = plt.subplots(figsize=(10,10))
    plt.title("New edges added after Committee Formation")
    plot_aggregated_trends(Y_nedge_list,TRIALS,iterations,closure_param_list,f,ax)
    plt.show()
    f,ax = plt.subplots(figsize=(10,10))
    plt.title("(Cumulative)New edges added after Committee Formation")
    plot_aggregated_trends(Y_nedge_list,TRIALS,iterations,closure_param_list,f,ax,cumulative=True)
    plt.show()
```
